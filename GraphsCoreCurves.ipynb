```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_graph = pd.read_csv("game_design_curves_model.csv")
df_graph
```

```python


# Melt the dataframe for seaborn compatibility
df_melted = df_graph.melt(id_vars='Milestone', var_name='Curve Type', value_name='Value')

# Plot all curves
plt.figure(figsize=(14, 8))
sns.lineplot(data=df_melted, x='Milestone', y='Value', hue='Curve Type', marker='o')
plt.xticks(rotation=45)
plt.title("Game Design Curves Over Progression Milestones")
plt.grid(True)
plt.tight_layout()
plt.show()
```
